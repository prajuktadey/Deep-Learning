# Count Vectorization

Count vectorization created a document-term matrix where the entry of each cell will be a count of the number of times that word occurred in that document.

## Read in text

```python
import pandas as pd
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

## Create function to remove punctuation, tokenize, remove stopwords, and stem

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords] #stemming words which are not there in stopwords
    return text
```

## Apply CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer=clean_text) #pass the function which is used to clean the text
X_counts = count_vect.fit_transform(data['body_text'])
#if we just the fit the data, it won't convert the data into feature vectors
#fit_transform is used to convert the data into feature vectors

print(X_counts.shape) #how many text messges we have
```

## Apply CountVectorizer to smaller samples

```python
data_sample = data[0:20]

count_vect_sample = CountVectorizer(analyzer=clean_text)
X_counts_sample = count_vect_sample.fit_transform(data_sample['body_text'])
print(X_counts_sample.shape)
```

## Vectorizers output sparse matrices

_**Sparse Matrix**: A matrix in which most entries are 0. In the interest of efficient storage, a sparse matrix will be stored by only storing the locations of the non-zero elements._
