Color may be encoded in variety of ways like RGB, RGBA, CMYK, or HSL. We'll use RGB encoding, where each pixel is represented as a vector with 3 channels: Red, Green, and Blue. RGB is usuallu encoded as a 24-bitmap, meaning that each pixel takes up 24 bits, or 3 bytes. Ecah byte represents a color channel, so every value must be an unsigned number between 0 and 255, where 0 means 0% intensity of its color, and 255 means 100% intensity.

```python
#creating a simple 3x3 image
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = 'retina'

img = np.array([[[255,   0,   0], [  0, 255,   0], [  0,   0, 255]],   #red, green, blue
                [[  0, 255, 255], [255,   0, 255], [255, 255,   0]],   #cyan, magenta, yellow
                [[  0,   0,   0], [255, 255, 255], [127, 127, 127]]])  #black, white, gray 50%

plt.imshow(img)
```

```python
#changing the lower left pixel from black to orange

img[2, 0, 0] = 255 #assign 100% to red
img[2, 0, 1] = 200 #assign about 80% to green

plt.imshow(img[2: ,:1]) #show pixel at row 2 and column 0
```
