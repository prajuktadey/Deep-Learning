### Prepare Input Data for Deep Learning

Perform the following steps for preparing data

1. Load data into a pandas dataframe
2. Convert the dataframe to a numpy array
3. Scale the feature dataset
4. Use one-hot-encoding for the target variable
5. Split into training and test datasets

```python
import pandas as pd
import os
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#Load Data and review content
iris_data = pd.read_csv("iris.csv")

print("\nLoaded Data :\n------------------------------------")
print(iris_data.head())
```

```python
#Use a Label encoder to convert String to numeric values 
#for the target variable

#the target variable species is a text field, it needs to be converted to a numeric representation
#to be used in deep learning

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
iris_data['Species'] = label_encoder.fit_transform(
                                iris_data['Species'])
```

```python
#Convert input/data frame to a numpy array
#nupy array is the preferred input format for keras deep learning
np_iris = iris_data.to_numpy()
```

```python
#Separate feature and target variables
X_data = np_iris[:,0:4] #4 columns
Y_data=np_iris[:,4] #1 column
```

```python
print("\nFeatures before scaling :\n------------------------------------")
print(X_data[:5,:])
print("\nTarget before scaling :\n------------------------------------")
print(Y_data[:5])
```

```python
#Create a scaler model that is fit on the input data.
scaler = StandardScaler().fit(X_data)
```

```python
#Scale the numeric feature variables
X_data = scaler.transform(X_data)

#Convert target variable as a one-hot-encoding array
Y_data = tf.keras.utils.to_categorical(Y_data,3)
```

```python
print("\nFeatures after scaling :\n------------------------------------")
print(X_data[:5,:])
print("\nTarget after one-hot-encoding :\n------------------------------------")
print(Y_data[:5,:])
```

```python
#Split training and test data
X_train,X_test,Y_train,Y_test = train_test_split( X_data, Y_data, test_size=0.10)
```

```python
print("\nTrain Test Dimensions:\n------------------------------------")
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

We are not splitting for validation because keras has a feature to create a validation data set from the tarining data set.

### Creating a Model

Creating a model in Keras requires defining the following

1. Number of hidden layers
2. Number of nodes in each layer
3. Activation functions
4. Loss Function & Accuracy measurements

```python
from tensorflow import keras

#Number of classes in the target variable
NB_CLASSES=3
```

```python
#Create a sequencial model in Keras
model = tf.keras.models.Sequential()
```

```python
#Add the first hidden layer
model.add(keras.layers.Dense(128,                    #Number of nodes
                             input_shape=(4,),       #Number of input variables
                              name='Hidden-Layer-1', #Logical name
                              activation='relu'))    #activation function
```

```python
#Add a second hidden layer
model.add(keras.layers.Dense(128,
                              name='Hidden-Layer-2',
                              activation='relu'))
```

```python
#Add an output layer with softmax activation
model.add(keras.layers.Dense(NB_CLASSES,
                             name='Output-Layer',
                             activation='softmax'))
```

```python
#Compile the model with loss & metrics
model.compile(loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Print the model meta-data
model.summary()
```

### Training and evaluating the Model

Training the model involves defining various training models and then perform 
forward and back propagation.

```python
#Make it verbose so we can see the progress
VERBOSE=1
```

```python
#Set Batch size
BATCH_SIZE=16
#Set number of epochs
EPOCHS=10
#Set validation split. 20% of the training data will be used for validation
#after each epoch
VALIDATION_SPLIT=0.2
```

```python
print("\nTraining Progress:\n------------------------------------")
```

```python
#Fit the model. This will perform the entire training cycle, including
#forward propagation, loss computation, backward propagation and gradient descent.
#Execute for the specified batch sizes and epoch
#Perform validation after each epoch 
history=model.fit(X_train,
          Y_train,
          batch_size=BATCH_SIZE,
          epochs=EPOCHS,
          verbose=VERBOSE,
          validation_split=VALIDATION_SPLIT)
```

```python
print("\nAccuracy during Training :\n------------------------------------")
import matplotlib.pyplot as plt
```

```python
#Plot accuracy of the model after each epoch.
pd.DataFrame(history.history)["accuracy"].plot(figsize=(8, 5))
plt.title("Accuracy improvements with Epoch")
plt.show()
```

```python
#Evaluate the model against the test dataset and print results
print("\nEvaluation against Test Dataset :\n------------------------------------")
model.evaluate(X_test,Y_test)
```

### Saving and Loading Models

The training and inference environments are usually separate. Models need to be saved after they are validated. They are then loaded into the inference environments for actual prediction

```python
#Saving a model
    
model.save("iris_save")
    
#Loading a Model 
loaded_model = keras.models.load_model("iris_save")

#Print Model Summary
loaded_model.summary()
```

### Predictions with Deep Learning Models

```python
#Raw prediction data
prediction_input = [[6.6, 3. , 4.4, 1.4]] #column header values 

#Scale prediction data with the same scaling model
scaled_input = scaler.transform(prediction_input)

#Get raw prediction probabilities
raw_prediction = model.predict(scaled_input)
print("Raw Prediction Output (Probabilities) :" , raw_prediction) # probable classes

#Find prediction
prediction = np.argmax(raw_prediction) #argmax is used to get the index of the highest probability
print("Prediction is ", label_encoder.inverse_transform([prediction]))
# we use reverse label encoder to get the string name
```
